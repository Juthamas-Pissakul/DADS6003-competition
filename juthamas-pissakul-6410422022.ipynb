# 3. Training data - Cross validation

```python
X= df_train.drop('y', axis = 1)
y= df_train['y']
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size = 0.3, random_state=123)
```

```python
X.head(10)
```

```python
logreg = LogisticRegression()
dt = DecisionTreeClassifier()
knn = KNeighborsClassifier()
forest = RandomForestClassifier()
grad_clf = GradientBoostingClassifier()
QDA = QuadraticDiscriminantAnalysis()
LDA = LinearDiscriminantAnalysis()
MPL = MLPClassifier()

  
dictM = {0: 'LogisticRegression',1: 'DecisionTreeClassifier', 2:'KNeighborsClassifier'
           , 3:'RandomForestClassifier', 4:'GradientBoostingClassifier', 5:'QDA', 6: 'LDA',7: 'MPL'}    
models = [logreg, dt, knn, forest, grad_clf, QDA, LDA, MPL]

for i,model in enumerate(models):
    print("\n{} Test Accuracy_CV: {}".format(dictM[i],cross_val_score(model, X, y, cv=10, scoring ='accuracy').mean()))
```

#### choose QuadraticDiscriminantAnalysis 

```python
QDA = QuadraticDiscriminantAnalysis()
QDA.fit( X_train,y_train)
pred = QDA.predict(X_test)
print(classification_report(y_test, pred))
```
